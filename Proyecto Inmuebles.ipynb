```python
import seaborn as sns
import matplotlib.pyplot as plt

# Definir el color map
n_color = 13
cMap = plt.cm.get_cmap("bwr", lut=n_color)

# Definir el heatmap
fig = plt.figure(figsize=(10, 8))
plt.title('Idealista Correlations')

# Ajustar los valores de vmin y vmax para reflejar todo el rango de correlaciones
h_map = sns.heatmap(
    idealista_df[variables_corr].corr().values, 
    vmin=-1, vmax=1, 
    cmap=cMap,
    annot=True, 
    xticklabels=list(idealista_df[variables_corr].columns),
    yticklabels=list(idealista_df[variables_corr].columns)
)
plt.xticks(rotation=90)

# Poner ticks en la barra de calor
cbar = h_map.collections[0].colorbar
l_ticks = [k/10. for k in range(-10, 11, 2)]  # Ticks ajustados para cubrir el rango [-1, 1]
cbar.set_ticks(l_ticks)

bottom, top = h_map.get_ylim()
h_map.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```

### Modeling


```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = idealista_df.drop(['PRICE', 'ASSETID'], axis = 1)
y = idealista_df['PRICE']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)
X_train
```

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
rf_predictions = rf_model.predict(X_test)
lr_predictions = lr_model.predict(X_test)

rf_mse = mean_squared_error(y_test, rf_predictions)
lr_mse = mean_squared_error(y_test, lr_predictions)

print("Random Forest MSE: ", np.sqrt(rf_mse))
print("Linear Regression MSE: ", np.sqrt(lr_mse))

print('\nSample Predictions: ')
num_samples = 10
for i in range(num_samples):
    print(f" - Random Forest: Predicted:  , {rf_predictions[i]}, Actual: {y_test.iloc[i]}")
    print(f" - Linear Regression: Predicted:  , {lr_predictions[i]}, Actual: {y_test.iloc[i]}")
    print()
```
